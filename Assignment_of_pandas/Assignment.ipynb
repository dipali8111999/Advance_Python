## Hitters dataset

```python
import pandas as pd
df = pd.read_csv('hitters.csv')
```

```python
pd.set_option('display.max_columns',25)
pd.set_option('display.max_rows',5)
```

```python
df.head()
```

#### Question 1: Find out no of missing values in each column

```python
df.isna().sum()
```

#### Question 2 : Find of people who have batted for more than 400 matches

```python
df[df['AtBat']>400][['Name','AtBat']]
```

#### Question 3 : Find out the name of the player who has hit max home runs

```python
mx_runs = df.max()['CHmRun']
mx_runs
```

```python
df[df['CHmRun']==mx_runs][['Name','CHmRun']]
```

#### Question 4 : Find out league wise total number of players

```python
df.groupby(by='League').count()[['Name']]
```

```python
df['League'].value_counts()
```

#### Question 5 : Find out division wise total amount of salary invested

```python
df.groupby(by='Division').sum()['Salary']
```

#### Question 6 : Find out five most senior players.

```python
df.sort_values(by='Years', ascending=False).head(5)[['Name', 'Years']]
```

## Cars93 Datasets

```python
data = pd.read_csv('Cars93.csv')
```

```python
pd.set_option('display.max_columns',100)
pd.set_option('display.max_rows',30)
```

```python
data.head()
```

```python
data.info()
```

```python
s = data.isna().sum()
s
```

```python
s[s>0]
```

```python
md = data['AirBags'].mode()[0]
data['AirBags'] = data['AirBags'].fillna('None')
```

```python
data.dtypes[['Rear.seat.room', 'Luggage.room']]
```

```python
mn = data[['Rear.seat.room', 'Luggage.room']].mean()
data[['Rear.seat.room', 'Luggage.room']] = data[['Rear.seat.room', 'Luggage.room']].fillna(mn)
```

```python
data.isna().sum()
```

```python
data.columns
```

#### Question 1 : Find out car having highest Mileage in city

```python
high_mpg = data['MPG.city'].max()
```

```python
data[data['MPG.city']==high_mpg][['Manufacturer', 'MPG.city']]
```

#### Question 2 : Find out models of cars which are having Price between 17 and 25

```python
data[(data['Price']>=17) & (data['Price']<=25)][['Model', 'Manufacturer','Price']]
```

#### Question 3 : Find out top 7 most fuel economic cars for highway

```python
data.sort_values(by='MPG.highway', ascending=False)[['Manufacturer','MPG.highway']].head(7)
```

```python
data['AirBags'].unique()
```

#### Question 4 : Find out Models of cars having no AirBags and which are of compact type whose price is below 20.

```python
data[(data['Price']<20)&(data['AirBags']=='None')&(data['Type']=='Compact')]
```

#### Question 5 : Find out top 7 most fuel economic cars for city, show results in sorted manner based on Horsepower

```python
p = data.sort_values('MPG.city', ascending=False).head(7)[['Horsepower','Manufacturer', 'Model',]]
p
```

```python
p.sort_values('Horsepower')
p.index = range(1,8,1)
```
